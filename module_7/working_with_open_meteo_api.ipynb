Scratchpad for exploring the OpenMeteo API.

```python
import httpx
def get_weather_forecast(lat: float, lon: float) -> float:
    base_url = "https://api.open-meteo.com/v1/forecast"
    params = {
        "latitude": lat,
        "longitude": lon,
        "hourly": ["temperature_2m", "snowfall"] # current: for current weather
    }
    expected_weather = httpx.get(base_url, params=params)
    return expected_weather
```

```python
res = get_weather_forecast(39.09, -120.03)
res
```

```python
res.json()['hourly']
```

```python
res.json()['hourly']['temperature_2m'][0] # predicted temperature 
```

```python
res.json()['hourly']['snowfall'][0] # predicted snowfall
```
